```
age_unique=sorted(data.age.unique())
age_diabetes_values=data.groupby('age')['diabetes'].count().values
mean_diabetes=[]
for i,age in enumerate(age_unique):
    mean_diabetes.append(sum(data[data['age']==age].diabetes)/age_diabetes_values[i])
```

```
#data_sorted=data.sort_values(by='Age',ascending=True)
plt.figure(figsize=(10,5))
sns.pointplot(x=age_unique,y=mean_diabetes,color='red',alpha=0.8)
plt.xlabel('Age',fontsize = 15,color='blue')
plt.xticks(rotation=45)
plt.ylabel('Diabetes',fontsize = 15,color='blue')
plt.title('Age vs Diabetes',fontsize = 15,color='blue')
plt.grid()
plt.show()
```

```
age_unique=sorted(data.age.unique())
age_valvular_disease_values=data.groupby('age')['valvular_disease'].count().values
mean_valvular_disease=[]
for i,age in enumerate(age_unique):
    mean_valvular_disease.append(sum(data[data['age']==age].valvular_disease)/age_valvular_disease_values[i])
```
